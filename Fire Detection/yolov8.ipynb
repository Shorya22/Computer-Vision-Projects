```python
import cv2
from ultralytics import YOLO

# Load a pretrained YOLOv8 model
model = YOLO('Fire_Detection.pt')

# Open a video file or start a video stream
cap = cv2.VideoCapture(0)  # replace with a video file path if not using a webcam

while cap.isOpened():
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Define the ROI coordinates
    roi_start_x, roi_start_y = 40, 50
    roi_end_x, roi_end_y = 300, 260

    # Draw the rectangle around the ROI on the frame
    cv2.rectangle(frame, (roi_start_x, roi_start_y), (roi_end_x, roi_end_y), (0, 255, 0), 2)

    # Extract the region of interest (ROI)
    roi = frame[roi_start_y:roi_end_y, roi_start_x:roi_end_x]

    # Run inference on the ROI
    results = model(roi, conf=0.8)

    # Plot results on the ROI
    for r in results:
        roi = r.plot()  # Plot detections on ROI

    # Replace the ROI in the original frame with the processed ROI
    frame[roi_start_y:roi_end_y, roi_start_x:roi_end_x] = roi

    # Display the resulting frame
    cv2.imshow('frame', frame)
    
    # Press 'q' on keyboard to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
